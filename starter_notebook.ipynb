# 🏦 THE LIFESTYLE ARCHITECT

University students often struggle to maintain a budget because their income is irregular. Part-time work, casual shifts, or allowances and their expenses are highly variable textbooks at the start of the semester, weekly transit, social events, and food. Without clear visibility, students frequently overspend in social categories and run out of money for essential living expenses. The lifestyle architect solves this by letting students bank CSV export.

## **Step 3 ** WORK THE PROBLEM BY HAND

CALCULATIONS

Total Outflow = Sum of all individual positive transations
Total inflow = Sum of all individual negetive transactions
Net Expenses = Total Outflow - Total inflow
Category Percentage = (Total Spent in one Category/Total Outflow)*100
Average Transaction Amount = Total Money spent/Total Number of Transactions

````python
# 🤖 AI Collaboration Prompt used: "Create a function to load a personal finance CSV transaction data with Date, Amount, Category, and Description columns.
# Handle Australian dollar signs, commas, and negative signs in the Amount column.
# If Category or Description are missing, fill them with 'Uncategorized'. Include clear, user-friendly error messages."

import pandas as pd
import numpy as np

def load_and_clean_transaction_data(file_path):
    """
    Loads and cleans student transaction data for The Lifestyle Architect tracker.
    Handles formatting issues like '$' symbols, missing values, and type conversions.
    """
    try:
        # Load the CSV file into a Pandas DataFrame
        # (file_path can be a string path or a Gradio file object)
        if isinstance(file_path, str):
            df = pd.read_csv(file_path)
        else:
            df = pd.read_csv(file_path.name)

        # 1. Validate Columns exist
        required_columns = ['Date', 'Amount', 'Category', 'Description']
        for col in required_columns:
            if col not in df.columns:
                raise ValueError(f"Missing required column: '{col}'. Please check your bank statement format.")

        # 2. Clean the 'Amount' column
        # Convert to string first to handle any mixed data types, then strip characters
        df['Amount'] = df['Amount'].astype(str).str.replace('$', '', regex=False)
        df['Amount'] = df['Amount'].str.replace(',', '', regex=False)
        df['Amount'] = df['Amount'].str.strip()

        # Convert to numeric float values
        df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

        # Drop rows where the Amount couldn't be parsed into a number
        df = df.dropna(subset=['Amount'])

        # 3. Handle Missing Text Values
        df['Category'] = df['Category'].fillna('Uncategorized').str.strip()
        df['Description'] = df['Description'].fillna('Unknown Merchant').str.strip()

        print("📊 Data successfully loaded and cleaned!")
        return df

    except Exception as e:
        print(f"❌ Error loading data: {e}")
        return None
````

````python
# =====================================================================
# 🏰 THE LIFESTYLE ARCHITECT — COMPLETE FINANCIAL ANALYZER SYSTEM
# =====================================================================

# =====================================================================
# 📦 IMPORTS
# =====================================================================

import os
import pandas as pd
import gradio as gr
from typing import Dict, Any, Optional

# Install missing package
!pip install hands_on_ai

# AI Imports
from hands_on_ai.chat import get_response
import hands_on_ai.rag as rag

# =====================================================================
# 🔐 ENVIRONMENT VARIABLES
# =====================================================================

os.environ['HANDS_ON_AI_SERVER'] = 'https://ollama.serveur.au'
os.environ['HANDS_ON_AI_MODEL'] = 'llama3.2'
os.environ['HANDS_ON_AI_API_KEY'] = 'isys2001-assignment-key'

# =====================================================================
# 📑 SAMPLE TRANSACTION DATA SETUP
# =====================================================================

print("📑 Generating sample transaction dataset...")
print("=" * 70)

messy_ledger_data = """Date,Amount,Category,Description
2026-03-01,$1200.00,Income,Part-time tutoring allowance
2026-03-02,450.00,Rent,Student Accommodation Fixed Weekly Pay
2026-03-03,$12.50,Groceries,Coles Supermarket purchase
2026-03-04,$6.80,Cafe,Campus Guild Cappuccino
2026-03-05,85.20,Utilities,Synergy Electricity Contribution
2026-03-06,$120.00,Entertainment,Concert Ticket booking
2026-03-07,$22.00,Transport,Transperth SmartRider Autoload
2026-03-08,,Food,Missing ledger item entry test case
"""

target_csv_filename = "student_bank_statement.csv"

with open(target_csv_filename, "w", encoding="utf-8") as csv_file:
    csv_file.write(messy_ledger_data.strip())

print(f"✅ CSV dataset saved successfully: {target_csv_filename}")

# =====================================================================
# ⚙️ FOUNDATION DATA PROCESSING FUNCTIONS
# =====================================================================

def load_and_sanitize_statement(file_path: str) -> Optional[pd.DataFrame]:

    try:

        if not os.path.exists(file_path):
            raise FileNotFoundError(f"File not found: {file_path}")

        df = pd.read_csv(file_path, skipinitialspace=True)

        required_schema = ['Date', 'Amount', 'Category', 'Description']

        if not all(col in df.columns for col in required_schema):
            raise KeyError(
                f"Dataset must contain: {required_schema}"
            )

        sanitized_df = df[required_schema].copy()

        # Clean Amount column
        sanitized_df['Amount'] = (
            sanitized_df['Amount']
            .astype(str)
            .str.replace('$', '', regex=False)
            .str.replace(',', '', regex=False)
            .str.strip()
        )

        sanitized_df['Amount'] = pd.to_numeric(
            sanitized_df['Amount'],
            errors='coerce'
        )

        # Remove invalid rows
        sanitized_df = sanitized_df.dropna(subset=['Amount'])

        # Clean categories
        sanitized_df['Category'] = (
            sanitized_df['Category']
            .fillna('Uncategorized')
            .astype(str)
            .str.strip()
            .str.title()
        )

        sanitized_df['Description'] = (
            sanitized_df['Description']
            .fillna('Unknown Transaction')
            .astype(str)
            .str.strip()
        )

        return sanitized_df.reset_index(drop=True)

    except Exception as error_context:

        print(f"❌ Processing Error: {error_context}")
        return None

# =====================================================================
# 💰 WALLET ANALYSIS ENGINE
# =====================================================================

def execute_wallet_allocation_audit(
    cleaned_df: pd.DataFrame
) -> Dict[str, Any]:

    # Separate income vs expenses properly
    income_records = cleaned_df[
        cleaned_df['Category'].str.lower() == 'income'
    ]

    expense_records = cleaned_df[
        cleaned_df['Category'].str.lower() != 'income'
    ]

    total_inflows = float(income_records['Amount'].sum())

    total_outflows = float(expense_records['Amount'].sum())

    # Essential categories
    essential_tags = [
        'Rent',
        'Groceries',
        'Utilities',
        'Transport',
        'Bills'
    ]

    category_aggregates = (
        expense_records
        .groupby('Category')['Amount']
        .sum()
    )

    essential_spend_total = 0.0
    discretionary_spend_total = 0.0

    category_summary_metrics = []

    for category_name, category_total in category_aggregates.items():

        if category_name in essential_tags:
            essential_spend_total += category_total
        else:
            discretionary_spend_total += category_total

        wallet_share_pct = (
            category_total / total_outflows * 100
        ) if total_outflows > 0 else 0

        category_summary_metrics.append({
            'Category': category_name,
            'Total Expended': round(float(category_total), 2),
            'Wallet Share %': round(wallet_share_pct, 1)
        })

    return {
        'net_outflows': total_outflows,
        'net_inflows': total_inflows,
        'essential_allocation': essential_spend_total,
        'discretionary_allocation': discretionary_spend_total,
        'summary_dataframe': pd.DataFrame(category_summary_metrics)
    }

# =====================================================================
# 📊 FINANCIAL INSIGHT ENGINE
# =====================================================================

def generate_financial_recommendations(
    financial_kpis: Dict[str, Any]
) -> str:

    outflows = financial_kpis.get('net_outflows', 0.0)
    inflows = financial_kpis.get('net_inflows', 0.0)
    essential = financial_kpis.get('essential_allocation', 0.0)
    discretionary = financial_kpis.get('discretionary_allocation', 0.0)
    summary_df = financial_kpis.get('summary_dataframe', pd.DataFrame())

    discretionary_pct = (
        discretionary / outflows * 100
    ) if outflows > 0 else 0

    net_savings_margin = inflows - outflows

    report = "# 🏰 Lifestyle Architect Financial Report\n\n"

    report += "## 📊 Spending Overview\n\n"

    report += (
        f"- 💵 Total Income: **${inflows:,.2f}**\n"
    )

    report += (
        f"- 💸 Total Expenses: **${outflows:,.2f}**\n"
    )

    report += (
        f"- 🧱 Essential Spending: **${essential:,.2f}**\n"
    )

    report += (
        f"- 🎯 Discretionary Spending: **${discretionary:,.2f}**\n\n"
    )

    report += (
        f"- 📈 Discretionary Wallet Share: "
        f"**{discretionary_pct:.1f}%**\n\n"
    )

    # Highest category
    if not summary_df.empty:

        highest_row = summary_df.loc[
            summary_df['Total Expended'].idxmax()
        ]

        report += (
            f"## 📌 Largest Spending Category\n\n"
            f"Your highest expense category is "
            f"**{highest_row['Category']}**, consuming "
            f"**{highest_row['Wallet Share %']}%** "
            f"of total expenses.\n\n"
        )

    # Risk assessment
    report += "## 🛑 Financial Risk Assessment\n\n"

    if discretionary_pct > 50:

        report += (
            "⚠️ **Lifestyle Leakage Crisis Detected**\n\n"
            "Your discretionary spending exceeds healthy "
            "student budgeting thresholds.\n\n"
        )

    elif discretionary_pct > 30:

        report += (
            "⚠️ **Moderate Spending Warning**\n\n"
            "Your flexible spending is slightly elevated.\n\n"
        )

    else:

        report += (
            "✅ **Healthy Budget Structure Detected**\n\n"
            "Your spending profile remains financially stable.\n\n"
        )

    # Savings section
    report += "## 💡 Recommendations\n\n"

    if net_savings_margin > 0:

        report += (
            f"1. ✅ Monthly surplus detected: "
            f"**${net_savings_margin:,.2f}**\n\n"
        )

        report += (
            "2. 🏦 Automatically transfer 20% "
            "into savings.\n\n"
        )

    else:

        report += (
            f"1. ❌ Deficit detected: "
            f"**${abs(net_savings_margin):,.2f}**\n\n"
        )

    report += (
        "3. 🍱 Meal prep instead of takeout.\n\n"
        "4. 🎓 Use student discounts wherever possible.\n\n"
        "5. 🚍 Use public transport concessions.\n"
    )

    return report

# =====================================================================
# 🤖 AI CHATBOT
# =====================================================================

def execution_architect_chatbot(user_query: str) -> str:

    if not user_query.strip():

        return (
            "👋 Ask me anything about budgeting, "
            "saving, or student finance!"
        )

    system_persona = (
        "You are The Lifestyle Architect, "
        "a friendly student financial coach. "
        "Give practical budgeting advice."
    )

    # Combine system persona and user query into a single prompt
    full_prompt = f"{system_persona}\n\n{user_query}"

    try:

        ai_response = get_response(
            prompt=full_prompt
        )

        return str(ai_response)

    except Exception as connection_fault:

        return (
            f"⚠️ AI Server Error: {connection_fault}"
        )

# =====================================================================
# 📚 RAG KNOWLEDGE BASE SYSTEM
# =====================================================================

def initialize_and_query_finance_rag(user_question: str):

    docs_directory = "student_knowledge_base"

    os.makedirs(docs_directory, exist_ok=True)

    knowledge_path = os.path.join(
        docs_directory,
        "student_welfare_guidelines.txt"
    )

    welfare_rules_content = """
THE LIFESTYLE ARCHITECT OFFICIAL STUDENT BUDGET MANUAL

RULE-101:
Students should follow the 50/30/20 budgeting rule.

RULE-102:
If discretionary spending exceeds 50%,
the student enters a Lifestyle Leakage Crisis.

RULE-103:
Students should maximize concessions,
discounts, and public transport savings.
"""

    with open(knowledge_path, "w", encoding="utf-8") as file:
        file.write(welfare_rules_content)

    try:

        # IMPORTANT:
        # Replace rag.query if your package uses another name
        retrieved_context = rag.query(
            query=user_question,
            docs_dir=docs_directory,
            k=1
        )

        if isinstance(retrieved_context, list):
            retrieved_context = "\n".join(
                [str(chunk) for chunk in retrieved_context]
            )

        system_persona = (
            "You are The Lifestyle Architect AI coach. "
            "Use ONLY the provided context."
        )

        augmented_prompt = f"""
{system_persona}

Document Context:
{retrieved_context}

Student Question:
{user_question}
"""

        ai_response = get_response(
            prompt=augmented_prompt
        )

        return (
            f"## 📚 Retrieved Knowledge\n\n"
            f"{retrieved_context}\n\n"
            f"---\n\n"
            f"## 🤖 AI Answer\n\n"
            f"{ai_response}"
        )

    except Exception as error:

        return f"❌ RAG System Error: {error}"

# =====================================================================
# 🧪 VALIDATION TESTS
# =====================================================================

print("\n🧪 Running validation tests...")
print("=" * 70)

processed_ledger = load_and_sanitize_statement(
    target_csv_filename
)

assert processed_ledger is not None

assert len(processed_ledger) == 7

assert processed_ledger['Amount'].dtype == 'float64'

financial_kpis = execute_wallet_allocation_audit(
    processed_ledger
)

assert round(
    financial_kpis['essential_allocation'],
    2
) == 569.70

print("✅ ALL VALIDATION TESTS PASSED")

# =====================================================================
# 🌐 GRADIO UI FUNCTIONS
# =====================================================================

def run_financial_audit():

    cleaned_df = load_and_sanitize_statement(
        target_csv_filename
    )

    if cleaned_df is None:
        return "❌ Failed to process statement."

    kpis = execute_wallet_allocation_audit(
        cleaned_df
    )

    report = generate_financial_recommendations(
        kpis
    )

    return report

def run_rag_query(user_question):

    return initialize_and_query_finance_rag(
        user_question
    )

# =====================================================================
# 🌐 GRADIO APPLICATION
# =====================================================================

with gr.Blocks(theme=gr.themes.Soft()) as app:

    gr.Markdown("""
    # 🏰 Lifestyle Architect Financial Intelligence Platform

    ### Features
    - 📊 Transaction Analysis
    - 💰 Wallet Allocation Audit
    - 🤖 AI Financial Coach
    - 📚 RAG Knowledge System
    - 🎓 Student Budget Guidance
    """)

    # =========================================================
    # TAB 1
    # =========================================================
    with gr.Tab("📊 Financial Audit"):

        gr.Markdown(
            "Run a complete financial analysis."
        )

        analyze_btn = gr.Button(
            "Run Financial Audit"
        )

        audit_output = gr.Markdown()

        analyze_btn.click(
            fn=run_financial_audit,
            outputs=audit_output
        )

    # =========================================================
    # TAB 2
    # =========================================================
    with gr.Tab("🤖 AI Financial Coach"):

        chatbot_input = gr.Textbox(
            label="Ask The Lifestyle Architect",
            placeholder=(
                "How can I reduce my spending?"
            )
        )

        chatbot_output = gr.Markdown()

        send_btn = gr.Button("Send")

        send_btn.click(
            fn=execution_architect_chatbot,
            inputs=chatbot_input,
            outputs=chatbot_output
        )

    # =========================================================
    # TAB 3
    # =========================================================
    with gr.Tab("📚 RAG Knowledge Assistant"):

        rag_input = gr.Textbox(
            label="Ask a financial policy question",
            placeholder=(
                "What happens if my discretionary "
                "spending exceeds 50%?"
            )
        )

        rag_output = gr.Markdown()

        rag_btn = gr.Button(
            "Run Knowledge Retrieval"
        )

        rag_btn.click(
            fn=run_rag_query,
            inputs=rag_input,
            outputs=rag_output
        )
def rag_query(question: str):

    try:
        context = rag.query(
            query=question,
            docs_dir="knowledge_base",
            k=1
        )

        if isinstance(context, list):
            context = "\n".join(context)

        full_prompt = f"""
        You are The Lifestyle Architect AI.

        Use ONLY the context below.

        Context:
        {context}

        Question:
        {question}
        """

        response = get_response(full_prompt)

        return f"""
### 📚 Context
{context}

### 🤖 Answer
{response}
"""

    except Exception as e:
        return f"RAG Error: {e}"

# =====================================================================
# 🚀 LAUNCH APPLICATION
# =====================================================================

print("\n🚀 Launching Gradio Application...")
print("=" * 70)

app.launch(debug=True)
````

## Custom Financial Tools

````python
# 🤖 AI Collaboration: Custom Tool Development
# Ask AI to help you create useful financial calculators

!pip install hands-on-ai
from hands_on_ai import agent

def create_savings_calculator_tool():
    """
    Create a custom savings goal calculator as an agent tool

    🤖 AI Collaboration Prompt:
    "Create a savings goal calculator function that takes current savings,
    monthly contribution, and target amount, then calculates time to reach
    goal. Format output for user-friendly display."
    """
    def calculate_savings_goal_timeline(
        current_savings: float,
        monthly_contribution: float,
        target_amount: float
    ) -> str:
        """
        Calculates how long it will take a user to reach a savings goal
        based on current savings and monthly contribution patterns.
        """

        # ----------------------------------------------------------------
        # INPUT VALIDATION
        # ----------------------------------------------------------------
        if current_savings < 0:
            return "❌ Current savings cannot be negative."

        if monthly_contribution <= 0:
            return "❌ Monthly contribution must be greater than $0."

        if target_amount <= 0:
            return "❌ Target amount must be greater than $0."

        if current_savings >= target_amount:
            return (
                "🎉 Congratulations! "
                "You have already reached your savings goal."
            )

        # ----------------------------------------------------------------
        # CALCULATION ENGINE
        # ----------------------------------------------------------------
        remaining_amount = target_amount - current_savings

        months_required = remaining_amount / monthly_contribution

        # Round upward because partial months still require saving
        months_required = int(months_required) + (
            1 if remaining_amount % monthly_contribution != 0 else 0
        )

        years = months_required // 12
        months = months_required % 12

        # ----------------------------------------------------------------
        # USER-FRIENDLY REPORT OUTPUT
        # ----------------------------------------------------------------
        report = "\n🏦 SMART SAVINGS GOAL CALCULATOR\n"
        report += "=" * 60 + "\n\n"

        report += f"💰 Current Savings: ${current_savings:,.2f}\n"
        report += f"📥 Monthly Contribution: ${monthly_contribution:,.2f}\n"
        report += f"🎯 Savings Goal Target: ${target_amount:,.2f}\n\n"

        report += f"📊 Remaining Amount Needed: ${remaining_amount:,.2f}\n"

        # Timeline formatting
        if years > 0:
            report += (
                f"⏳ Estimated Time to Goal: "
                f"{years} year(s) and {months} month(s)\n\n"
            )
        else:
            report += (
                f"⏳ Estimated Time to Goal: "
                f"{months_required} month(s)\n\n"
            )

        # Motivational insight layer
        if months_required <= 6:
            report += (
                "✅ Excellent progress trajectory! "
                "Your savings behaviour is highly efficient.\n"
            )

        elif months_required <= 24:
            report += (
                "📈 Stable savings momentum detected. "
                "Maintaining consistency will achieve your target comfortably.\n"
            )

        else:
            report += (
                "⚠️ Long-term savings horizon detected. "
                "Consider increasing monthly contributions or reducing "
                "discretionary spending leakage to accelerate your goal.\n"
            )

        return report
````
